1. Write a command to add a new column ProfitRank to the Delta Lake table that ranks each row within its respective 'Region' based on TotalProfit in descending order.

```
%sql
SELECT *, RANK() OVER (PARTITION BY Region ORDER BY TotalProfit DESC) AS ProfitRank
FROM Orders
```

<b>
 2. How would you calculate the cumulative TotalRevenue for each 'Country' in the Delta Lake table? Provide a SQL command to achieve this.

```
%sql
SELECT Country, SUM(TotalRevenue) OVER (PARTITION BY Country ORDER BY OrderID) as CumulativeTotalRevenue
FROM Orders
```

 3. Write a command to find the average TotalProfit for each 'ItemType' over the entire data. Then, add a new column ProfitDifference to the table, which shows the difference between the TotalProfit of each row and the average TotalProfit of its 'ItemType'.

```
%sql
SELECT *, TotalProfit - AVG(TotalProfit) OVER (PARTITION BY ItemType) as ProfitDifference
FROM Orders
```

 4. How would you partition the data by 'Region' and calculate the running total of UnitsSold within each partition? Write the SQL query for this.

```
%sql
SELECT *, SUM(UnitsSold) OVER (PARTITION BY Region ORDER BY OrderID) as RunningTotalUnitsSold
FROM Orders
```

 5. Create a SQL query that adds a new column SalesPercentile to the table that shows the percentile rank of TotalRevenue for each 'ItemType' within its respective 'Region'.

```
%sql
SELECT *, PERCENT_RANK() OVER (PARTITION BY ItemType, Region ORDER BY TotalRevenue) as SalesPercentile
FROM Orders
```

 6. Write a SQL query that identifies the 'Region' and 'Country' where the TotalProfit has the highest standard deviation.

```
%sql
SELECT Region, Country, STDDEV(TotalProfit) OVER (PARTITION BY Region, Country) as StdDevTotalProfit
FROM Orders
```

 7. Write a SQL query that calculates the ratio of each row's TotalRevenue to the sum TotalRevenue of its 'Region' and 'ItemType'.

```
%sql
SELECT *, TotalRevenue / SUM(TotalRevenue) OVER (PARTITION BY Region, ItemType) as RevenueRatio
FROM Orders
```

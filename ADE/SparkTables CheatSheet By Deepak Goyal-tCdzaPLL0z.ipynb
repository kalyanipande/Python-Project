<b>  Create an internal Parquet table

```
spark.sql("CREATE TABLE internal_table (id INT, name STRING) USING parquet")


```

<b>  Create an internal CSV table

```
spark.sql("CREATE TABLE internal_table (id INT, name STRING) USING csv OPTIONS (path 'path/to/csv', header 'true')")


```

<b>  Create an internal JSON table

```
spark.sql("CREATE TABLE internal_table AS SELECT * FROM existing_table")

```

<b>  Create an internal table with partitioning

```
spark.sql("CREATE TABLE internal_table (id INT, name STRING) PARTITIONED BY (date DATE) USING parquet")

```

<b>  Create an external Parquet table

```
spark.sql("CREATE EXTERNAL TABLE external_table (id INT, name STRING) USING parquet LOCATION 'path/to/parquet'")

```

<b>  Create an external CSV table

```
spark.sql("CREATE EXTERNAL TABLE external_table (id INT, name STRING) USING csv OPTIONS (path 'path/to/csv', header 'true') LOCATION 'path/to/csv'")



```

<b>  Create an external JSON table

```
spark.sql("CREATE EXTERNAL TABLE external_table (id INT, name STRING) USING json OPTIONS (path 'path/to/json') LOCATION 'path/to/json'")
```

md
<b>  Create an external table from select query

```
spark.sql("CREATE EXTERNAL TABLE external_table AS SELECT * FROM existing_table LOCATION 'path/to/data'")
```

md
<b>  Create an external table with partitioning

```
spark.sql("CREATE EXTERNAL TABLE external_table (id INT, name STRING) PARTITIONED BY (date DATE) USING parquet LOCATION 'path/to/parquet'"))

```

md
<b>  Load data into a table

```
spark.sql("LOAD DATA INPATH 'path/to/data' INTO TABLE table_name")

```

d
<b>  Add a partition to a table

```
spark.sql("ALTER TABLE table_name ADD PARTITION (date='2023-05-17') LOCATION 'path/to/data'")

```

d
<b>  Drop a partition from a table

```
spark.sql("ALTER TABLE table_name DROP PARTITION (date='2023-05-17')")

```

d
<b>  Describe a table

```
spark.sql("DESCRIBE TABLE table_name")

```

d
<b>  Refresh table metadata

```
spark.sql("REFRESH TABLE table_name")
```

d
<b>  Cache a table in memory

```
spark.sql("CACHE TABLE table_name")
```

d
<b>  Uncache a table from memory

```
spark.sql("UNCACHE TABLE table_name")
```

<b>  Create a temporary view

```
spark.sql("CREATE TEMPORARY VIEW temp_view AS SELECT * FROM table_name")
```

d
 <b>  Drop a temporary view

```
spark.sql("DROP VIEW IF EXISTS temp_view")
```

d
 <b>  Query a temporary view

```
spark.sql("SELECT * FROM temp_view")
```

d
 <b>  Create a global temporary view

```
spark.sql("CREATE GLOBAL TEMPORARY VIEW global_temp_view AS SELECT * FROM table_name")
```

d
 <b>  Drop a global temporary view

```
spark.sql("DROP GLOBAL VIEW IF EXISTS global_temp_view")
```

d
 <b>  Query a global temporary view

```
spark.sql("SELECT * FROM global_temp_view")
```

d
 <b>  Create a permanent view

```
spark.sql("CREATE VIEW view_name AS SELECT * FROM table_name")
```

d
 <b>  Create a permanent view in a specific database

```
spark.sql("CREATE VIEW database_name.view_name AS SELECT * FROM table_name")
```

md
 <b>  Drop a permanent view

```
spark.sql("DROP VIEW IF EXISTS view_name")
```

d
 <b>  Query a permanent view

```
spark.sql("SELECT * FROM view_name")
```
